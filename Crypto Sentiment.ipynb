```python
import tweepy as tw
import pandas as pd
from datetime import date
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import os
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from textblob import TextBlob
```

```python
auth = tw.OAuthHandler('*************************', '***********************************')
auth.set_access_token('**************************', '***********************************')
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
#Bitcoin

crypto_tweets = tw.Cursor(api.search_tweets,q = "Bitcoin OR BTC",lang = "en", tweet_mode="extended").items(1000)
crypto_tweets_list = [[tweet.created_at, tweet.user.location, tweet.user.name, tweet.full_text, tweet.id, tweet.user.followers_count] for tweet in crypto_tweets]
tweets_df_bitcoin = pd.DataFrame(crypto_tweets_list)
tweets_df_bitcoin.columns = ['Created at', 'Place', 'User', 'Text', 'ID', 'Followers count']
#tweets_df_bitcoin

#Ethereum
crypto_tweets = tw.Cursor(api.search_tweets,q = "ETH OR Ethereum",lang = "en", tweet_mode="extended").items(200)
crypto_tweets_list = [[tweet.created_at, tweet.user.location, tweet.user.name, tweet.full_text, tweet.id, tweet.user.followers_count] for tweet in crypto_tweets]
tweets_df_Ethereum = pd.DataFrame(crypto_tweets_list)
tweets_df_Ethereum.columns = ['Created at', 'Place', 'User', 'Text', 'ID', 'Followers count']
#tweets_df_Ethereum

#Doge
crypto_tweets = tw.Cursor(api.search_tweets,q = "DOGE",lang = "en", tweet_mode="extended").items(200)
crypto_tweets_list = [[tweet.created_at, tweet.user.location, tweet.user.name, tweet.full_text, tweet.id, tweet.user.followers_count] for tweet in crypto_tweets]
tweets_df_DOGE = pd.DataFrame(crypto_tweets_list)
tweets_df_DOGE.columns = ['Created at', 'Place', 'User', 'Text', 'ID', 'Followers count']
#tweets_df_DOGE

#Solana
crypto_tweets = tw.Cursor(api.search_tweets,q = "SOL OR Solana",lang = "en", tweet_mode="extended").items(200)
crypto_tweets_list = [[tweet.created_at, tweet.user.location, tweet.user.name, tweet.full_text, tweet.id, tweet.user.followers_count] for tweet in crypto_tweets]
tweets_df_solana = pd.DataFrame(crypto_tweets_list)
tweets_df_solana.columns = ['Created at', 'Place', 'User', 'Text', 'ID', 'Followers count']
#tweets_df_solana
```

```python
pd.set_option('display.max_colwidth', None)
data_bitcoin = tweets_df_bitcoin
data_eth = tweets_df_Ethereum
data_doge = tweets_df_DOGE
data_sol = tweets_df_solana
```

```python
#Bitcoin

new_data_btc = data_bitcoin.drop_duplicates('Text',keep='first') #delete the duplicates by dropping them and store the result value to a new variable
tweets = new_data_btc['Text']
tweets1 = []
for i in tweets:
    tweets1.append(i.replace('\n', ''))
    
tweets2 = []
for i in tweets1:
    tweets2.append(re.sub(r"(@[A-Za-z0–9_]+)|[^\w\s]|#|http\S+", "", i))
    
new_data_btc['Text'] = tweets2

stop_words = stopwords.words("english")
stemmer = SnowballStemmer("english", ignore_stopwords=True)
lemmatizer = WordNetLemmatizer()

texts = [[word.lower() for word in text.split()] for text in tweets2]
lemma_list_of_words = []
for i in range(0, len(texts)):
    l1 = texts[i]
    l2 = ' '.join([lemmatizer.lemmatize(word) for word in l1])
    lemma_list_of_words.append(l2)
    
stem_text = [stemmer.stem(i) for i in lemma_list_of_words]

x = []
sentiment = []
for s in stem_text:
    analysis = TextBlob(s)
    #print(analysis)
    x.append(analysis.sentiment.polarity)
    if analysis.sentiment.polarity > 0 and analysis.sentiment.polarity < 0.5 :
        sentiment.append('Positive')
    elif analysis.sentiment.polarity > 0.5:
        sentiment.append('Highly Positive')
    elif analysis.sentiment.polarity > -0.5 and analysis.sentiment.polarity < 0 :
        sentiment.append('Negative')
    elif analysis.sentiment.polarity < -0.5:
        sentiment.append('Highly Negaive')
        
    else:
        sentiment.append('Neutral')
        
new_data_btc['Polarity'] = x
new_data_btc['Sentiment'] = sentiment
new_data_btc['Cleaned_text'] = stem_text
new_data_btc['Symbol'] = 'BTC'

sorted_df_btc = new_data_btc.sort_values(by='Followers count', ascending=False)
sorted_df_btc



sorted_df_btc['Sentiment'].value_counts()
```

```python
weight=[]
for i in new_data_btc['Followers count']:
    if i<100:
        weight.append(0)
    elif i>=100 and i<1000:
        weight.append(0.125)
    elif i>=1000 and i<10000:
        weight.append(0.25)
    elif i>=10000 and i<100000:
        weight.append(0.5)
    elif i>=100000 and i<1000000:
        weight.append(0.75)
    else:
        weight.append(1)
```

```python
new_data_btc['weightage']=weight
```

```python
new_data_btc['score']=new_data_btc['Followers count']*new_data_btc['Polarity']*new_data_btc['weightage']
```

```python
recommendation=[]
for score in new_data_btc['score']:
    if score<=-5000:
        recommendation.append('Strong Sell')
    elif score<=-1000 and score>-5000:
        recommendation.append('Sell')
    elif score<1000 and score>-1000:
        recommendation.append('Hold')
    elif score>=1000 and score<5000:
        recommendation.append('Buy')
    else:
        recommendation.append('Strong Buy')
    
new_data_btc['remark']= recommendation
```
